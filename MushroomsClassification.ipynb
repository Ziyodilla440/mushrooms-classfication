**Mushroom dataset ustunlari va ularning tavsifi**  

1. **class** – qoʻziqorinning turi: zaharli (`p`) yoki zaharsiz (`e`). *Bu target (maqsad) ustun hisoblanadi.*  
2. **cap-shape** – qoʻziqorin shapkasi shakli:  
   - `b`: bell (qo‘ng‘iroq)  
   - `c`: conical (konussimon)  
   - `x`: convex (yumaloq)  
   - `f`: flat (tekis)  
   - `k`: knobbed (dumaloq tugunli)  
   - `s`: sunken (botiq)  

3. **cap-surface** – qoʻziqorin shapkasi yuzasi qanday koʻrinishda:  
   - `f`: fibrous (tolali)  
   - `g`: grooved (egik chiziqli)  
   - `y`: scaly (taroqli)  
   - `s`: smooth (silliq)  

4. **cap-color** – qoʻziqorin shapkasi rangi:  
   - `n`: brown (jigarrang)  
   - `b`: buff (to‘q sariq-jigarrang)  
   - `c`: cinnamon (dolchin rang)  
   - `g`: gray (kulrang)  
   - `r`: green (yashil)  
   - `p`: pink (pushti)  
   - `u`: purple (binafsha)  
   - `e`: red (qizil)  
   - `w`: white (oq)  
   - `y`: yellow (sariq)  

5. **bruises** – qoʻziqorin bosilganda dogʻ qoldiradimi?  
   - `t`: true (ha)  
   - `f`: false (yo‘q)  

6. **odor** – qoʻziqorin hidi:  
   - `a`: almond (bodom)  
   - `l`: anise (anis)  
   - `c`: creosote (kreozot – o‘ziga xos oʻtkir hid)  
   - `y`: fishy (baliq hidi)  
   - `f`: foul (yomon hid)  
   - `m`: musty (mog‘or hidi)  
   - `n`: none (hidsiz)  
   - `p`: pungent (keskin hid)  
   - `s`: spicy (achchiq hid)  

7. **gill-attachment** – qoʻziqorin plastinkalarining shapkaga bogʻlanish turi:  
   - `a`: attached (biriktirilgan)  
   - `d`: descending (pastga tushgan)  
   - `f`: free (erkin)  
   - `n`: notched (egik)  

8. **gill-spacing** – plastinkalar orasidagi masofa:  
   - `c`: close (yaqin)  
   - `w`: crowded (juda zich)  
   - `d`: distant (uzoq)  

9. **gill-size** – plastinkalarning o‘lchami:  
   - `b`: broad (keng)  
   - `n`: narrow (tor)  

10. **gill-color** – plastinkalarning rangi:  
    - `k`: black (qora)  
    - `n`: brown (jigarrang)  
    - `b`: buff (to‘q sariq-jigarrang)  
    - `h`: chocolate (shokolad rang)  
    - `g`: gray (kulrang)  
    - `r`: green (yashil)  
    - `o`: orange (to‘q sariq)  
    - `p`: pink (pushti)  
    - `u`: purple (binafsha)  
    - `e`: red (qizil)  
    - `w`: white (oq)  
    - `y`: yellow (sariq)  

11. **stalk-shape** – qoʻziqorin poyasi shakli:  
    - `e`: enlarging (yuqoriga qarab kengaygan)  
    - `t`: tapering (yuqoriga qarab toraygan)  

12. **stalk-root** – qoʻziqorin poyasining ildiz qismi:  
    - `b`: bulbous (kurtakli)  
    - `c`: club (to‘qmoqsimon)  
    - `u`: cup (chashka shaklida)  
    - `e`: equal (baravar)  
    - `z`: rhizomorphs (ildiz tolalar bilan bog‘langan)  
    - `r`: rooted (ildizlangan)  
    - `?`: nomaʼlum  

13. **stalk-surface-above-ring** – qoʻziqorin poyasining halqadan yuqoridagi yuzasi:  
    - `f`: fibrous (tolali)  
    - `y`: scaly (taroqli)  
    - `k`: silky (ipaksimon)  
    - `s`: smooth (silliq)  

14. **stalk-surface-below-ring** – qoʻziqorin poyasining halqadan pastdagi yuzasi:  
    - `f`: fibrous (tolali)  
    - `y`: scaly (taroqli)  
    - `k`: silky (ipaksimon)

15. **stalk-color-above-ring** – halqadan yuqoridagi poya rangi:  
    - `n`: brown (jigarrang)  
    - `b`: buff (to‘q sariq-jigarrang)  
    - `c`: cinnamon (dolchin rang)

16. **stalk-color-below-ring** – halqadan pastdagi poya rangi:

17. **veil-type** – qoʻziqorin pardasi turi (barcha qiymatlar `p`, yaʼni **partial** (qisman) boʻlgani uchun foydasiz ustun).  

18. **veil-color** – parda rangi:

```python
import pandas as pd
import numpy as np
import sklearn

url="https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/mushrooms.csv"
df = pd.read_csv(url)
df.head()
```

# **Analysis**

```python
#javobni shu yerda yozing.
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import tree
import seaborn as sns
from sklearn.metrics import (
    accuracy_score, 
    confusion_matrix, 
    precision_score, 
    recall_score, 
    f1_score,
    classification_report,
    ConfusionMatrixDisplay
)

import matplotlib.pyplot as plt
```

```python
print(df.shape)
print(df.info)
```

```python
df.drop('veil-type', axis=1, inplace=True)
df.isna().sum()
```

```python
unique_counts = df.nunique()
print(unique_counts)
```

```python
df['class'].value_counts() #balanced
```

**Encoder**

```python
x = df.drop("class", axis=1) 
y = df['class']

x_encoded = x.copy()

for col in x.columns:
    encoder = LabelEncoder()
    x_encoded[col] = encoder.fit_transform(x[col])

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)


```

**Visualisation**

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='class', data=df, palette='viridis')
plt.title('Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
print(df['class'].value_counts())
```

```python
categorical_cols = df.drop('class', axis=1).columns.tolist()
selected_viz_cols = [
    'cap-shape', 'cap-color', 'odor', 'gill-color', 'stalk-shape',
    'stalk-root', 'ring-type', 'spore-print-color', 'population', 'habitat'
]

plt.figure(figsize=(20, 18))
for i, col in enumerate(selected_viz_cols):
    plt.subplot(4, 3, i + 1)
    sns.countplot(x=col, data=df, palette='Spectral', order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 18))
influential_cols = [
    'odor', 'gill-color', 'spore-print-color', 'habitat', 'bruises', 'ring-type'
]

for i, col in enumerate(influential_cols):
    plt.subplot(3, 2, i + 1)
    sns.countplot(x=col, hue='class', data=df, palette='coolwarm', order=df[col].value_counts().index)
    plt.title(f'Impact of {col} on Class')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Class', labels=['Edible', 'Poisonous'])
plt.tight_layout()
plt.show()
```

# **Train/test split**

```python
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y_encoded, test_size=0.3, random_state=42)
```

**k-NN**

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_predict = knn.predict(x_test)
```

```python
# testing
accuracy = accuracy_score(y_test, y_predict)
precision = precision_score(y_test, y_predict)
recall = recall_score(y_test, y_predict)
f1 = f1_score(y_test, y_predict)

print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1:.4f}")

cm = confusion_matrix(y_test, y_predict)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.title("KNN Confusion Matrix")
plt.show()
```

**Decision Tree**

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(x_train, y_train)
y_predict = tree_model.predict(x_test)
```

```python
# testing
accuracy = accuracy_score(y_test, y_predict)
precision = precision_score(y_test, y_predict)
recall = recall_score(y_test, y_predict)
f1 = f1_score(y_test, y_predict)

print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1:.4f}")

cm = confusion_matrix(y_test, y_predict)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.title("Decision Tree Confusion Matrix")
plt.show()
```

**Random Forest**

```python
rf_model = RandomForestClassifier()
rf_model.fit(x_train, y_train)
y_predict = rf_model.predict(x_test)
```

```python
# testing
accuracy = accuracy_score(y_test, y_predict)
precision = precision_score(y_test, y_predict)
recall = recall_score(y_test, y_predict)
f1 = f1_score(y_test, y_predict)

print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1:.4f}")

cm = confusion_matrix(y_test, y_predict)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.title("Random Forest Confusion Matrix")
plt.show()
```

**Logistic Regression**

```python
lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)
y_predict = lr_model.predict(x_test)
```

```python
# testing
accuracy = accuracy_score(y_test, y_predict)
precision = precision_score(y_test, y_predict)
recall = recall_score(y_test, y_predict)
f1 = f1_score(y_test, y_predict)

print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1:.4f}")

cm = confusion_matrix(y_test, y_predict)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.title("Logistic Regression Confusion Matrix")
plt.show()
```

**Model Comparison**


```python
models = {
    'KNN': knn,
    'Decision Tree': tree_model,
    'Random Forest': rf_model,
    'Logistic Regression': lr_model,
}

results = []
for model_name, model in models.items():
    y_pred = model.predict(x_test)
    results.append({
        'Model': model_name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1-Score': f1_score(y_test, y_pred),
    })

results_df = pd.DataFrame(results).sort_values(
    by=['F1-Score', 'Accuracy', 'Precision', 'Recall'],
    ascending=False
).reset_index(drop=True)

print('Model comparison table:')
display(results_df.style.format({
    'Accuracy': '{:.4f}',
    'Precision': '{:.4f}',
    'Recall': '{:.4f}',
    'F1-Score': '{:.4f}'
}))

plt.figure(figsize=(10, 5))
melted = results_df.melt(id_vars='Model', value_vars=['Precision', 'Recall', 'F1-Score'])
sns.barplot(data=melted, x='Model', y='value', hue='variable', palette='Set2')
plt.title('Model Metrics Comparison (Precision, Recall, F1-Score)')
plt.xlabel('Model')
plt.ylabel('Score')
plt.ylim(0.0, 1.05)
plt.xticks(rotation=15)
plt.legend(title='Metric')
plt.tight_layout()
plt.show()

best_model = results_df.iloc[0]
print(f"Best model: {best_model['Model']}")
```
